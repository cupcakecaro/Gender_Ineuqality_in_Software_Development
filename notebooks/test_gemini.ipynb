## Gemini
See [here](https://aistudio.google.com/app/rate-limit?timeRange=last-28-days&project=gen-lang-client-0194085235) for an overview over rate limits for individual models.

```python
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)
model = "gemini-3-flash-preview"

# available models are "gemini-2.5-flash-lite", "gemini-2.5-flash", "gemini-3-flash-preview"
# try changing between them if you hit rate limits

response = client.models.generate_content(
    model=model, contents="Explain how AI works in a few words"
)
print(response.text)
```
